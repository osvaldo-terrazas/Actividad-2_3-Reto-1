```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
```

```python
#Carga desde un archivo .csv sin indice
df = pd.read_csv('Exportaciones.csv')
```

```python
#a) Las exportaciones de “iso3: alb”, y de “h_insurance” mayor a “0.5”
df_a = df[(df['iso3'] == 'alb') & (df['h_insurance'] > 0.5)]
df_a
```

```python
#b) Todas las exportaciones de “sector Gmatrix: insurance“ con “g_procfoods” mayor a “7”
df_b= df[(df['sector_GMatrix'] == 'insurance') & (df['g_procfoods'] > 7)]
df_b
```

```python
#c) Todas las exportaciones de “year: 1997 a 2005”
df_c= df[(df['year'] >= 1997) & (df['year'] <= 2005)]
df_c
```

```python
#d) Todas las exportaciones con “Vxsharefwd” mayor a “0.5” y con “Vxsharebwd” menor a “0.5”
df_d= df[(df['VXsharefwd'] > 0.5) & (df['VXsharebwd'] < 0.5)]
df_d
```

```python
#e) Las 100 exportaciones de “iso3: tha, kor, phl, ven y deu”, con “g_textiles” mayor a “1” y “g clothing” menor a “1”
df_e= df[(df['iso3'] == 'tha') | (df['iso3'] == 'kor') | (df['iso3'] == 'phl') | (df['iso3'] == 'ven') | (df['iso3'] == 'deu') & (df['g_textiles'] > 1) & (df['g_clothing'] < 1)]
df_e
```

```python
#f) Las 15 exportaciones de “year: 2004”, con “h_oservices” mayor a “10”
df_f= df[(df['year'] == 2004) & (df['h_oservices'] > 10)]
df_f
```

```python
#g) Todas las exportaciones que no sean de “sector_Gmatrix: machinery“
df_g= df[~(df['sector_GMatrix'] == 'machinery')]
df_g
```

```python
#h) Todas las exportaciones que no sean de “Dxshare “ mayor a “0.5”
df_h= df[~(df['DXshare'] > 0.5)]
df_h
```

```python
#i) Todas las exportaciones que sean de “h_oconsumer:“ igual a “0.02” o “h_obsict“ mayor que “10”
df_i= df[(df['h_oconsumer'] == 0.02) | (df['h_obsict'] > 10)]
df_i
```

```python
#Convertir filtros a archivos csv
df_a to_csv("df_a.csv")
df_b.to_csv("df_b.csv")
df_c.to_csv("df_c.csv")
df_d.to_csv("df_d.csv")
df_e.to_csv("df_e.csv")
df_f.to_csv("df_f.csv")
df_g.to_csv("df_g.csv")
df_h.to_csv("df_h.csv")
df_i.to_csv("df_i.csv")
```
